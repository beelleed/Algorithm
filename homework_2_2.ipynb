```python
import matplotlib.pyplot as plt
def fibonacci_count_f4(n, counter):
    if n == 4:
        counter['count'] += 1
    if n <= 1:
        return n
    else:
        return fibonacci_count_f4(n-1, counter) + fibonacci_count_f4(n-2, counter)

n_values = range(5, 51)
counts = []

for n in n_values:
    counter = {'count': 0}
    fibonacci_count_f4(n, counter)
    counts.append(counter['count'])

# Plot the results
plt.figure(figsize=(10, 5))
plt.plot(n_values, counts, label='Times F(4) Computed')
plt.xlabel('n')
plt.ylabel('Number of Computations of F(4)')
plt.title('Overlapping Subproblems in Fibonacci Calculation')
plt.legend()
plt.show()
```
