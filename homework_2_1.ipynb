```python
import time
import matplotlib.pyplot as plt
import json

def fibonacci_recursive(n):
    if n <= 1:
        return n
    else:
        return fibonacci_recursive(n-1) + fibonacci_recursive(n-2)

def fibonacci_dynamic(n):
    fib = [0] * (n + 1)
    fib[1] = 1
    for i in range(2, n + 1):
        fib[i] = fib[i - 1] + fib[i - 2]
    return fib[n]

# Prepare to save data
times_recursive = []
times_dynamic = []
n_values = range(1, 101)

# Run the calculations and save the data
for n in n_values:
    start_time = time.time()
    fibonacci_recursive(n)
    end_time = time.time()
    times_recursive.append(end_time - start_time)

    start_time = time.time()
    fibonacci_dynamic(n)
    end_time = time.time()
    times_dynamic.append(end_time - start_time)

    # Save the data to a JSON file periodically
    if n % 10 == 0:  # Save every 10 iterations
        with open('fibonacci_times.json', 'w') as f:
            json.dump({'times_recursive': times_recursive, 'times_dynamic': times_dynamic}, f)

# Plot the results
plt.figure(figsize=(10, 5))
plt.plot(n_values, times_recursive, label='Recursive Fibonacci')
plt.plot(n_values, times_dynamic, label='Dynamic Fibonacci')
plt.xlabel('n')
plt.ylabel('Execution Time (seconds)')
plt.title('Comparison of Fibonacci Computation Methods')
plt.legend()
plt.show()
```

```python
import json
import matplotlib.pyplot as plt

# 讀取 JSON 文件
with open('fibonacci_times.json', 'r') as file:
    data = json.load(file)

# 數據已經讀取，現在從數據中提取時間數據
times_recursive = data['times_recursive']
times_dynamic = data['times_dynamic']
n_values = list(range(1, len(times_recursive) + 1))

# 使用 Matplotlib 繪製折線圖
plt.figure(figsize=(10, 5))
plt.plot(n_values, times_recursive, label='Recursive Fibonacci')
plt.plot(n_values, times_dynamic, label='Dynamic Fibonacci')
plt.xlabel('n')
plt.ylabel('Execution Time (seconds)')
plt.title('Comparison of Fibonacci Computation Methods')
plt.legend()
plt.show()
```
